```python
random.seed(42)

def sample_audio(source, destination, *, seconds=30, format=None):
    """Save a sample from audio source from a random section of the song.
    
    PARAMS
    source: audio source file or path-like object
    destination: destination folder to save sample
    seconds: duration in seconds for sample
    """
    source = Path(source)
    destination = Path(destination)
    
    assert source.is_file()
    assert destination.is_dir()
    
    name = source.stem
    ext = source.suffix.strip('.')
    if format is None:
        format = ext
    
    audio = AudioSegment.from_file(source, format=ext)
    duration = audio.duration_seconds
    
    start = random.randint(0, duration - 30) * 1000
    stop = start + seconds * 1000
    
    sample = audio[start:stop]
    output = destination / f'{name}_{seconds}.{ext}'
    
    sample.export(output, format=format)
```
